```python
#3.   Bokeh 官方有提供 sample_data 給大家練習，gallery 豐富的範例都取自 sample_data，
#對比官方的資料格式就能輕鬆模仿應用，下載 股市資料
data = pd.read_csv('table_aapl.csv')
data.head()
```

```python
#data['date']=data['date'].apply(lambda x:str(x).time('%Y%m%d'))
data['date']=pd.to_datetime(data['date'], format='%Y%m%d')
data['date_str']=data['date'].apply(lambda x:str(x).split(' ')[0])
data.head()
```

```python
#4.   使用 HoverTool(游標滑過時顯示資料)；Click_policy (藉由標籤控制數值顯示)
source = ColumnDataSource(data=dict(
    date=data['date'],
    close=data['close'],
    high=data['high'],
    low=data['low'],
    date_str=data['date_str']
))

TOOLTIPS = [
    ("Date", '@date_str'),
    ('Price', '@close'),
    ("High", "@high"),
    ("Low", "@low")
    
]


p = figure(x_axis_type='datetime',tooltips=TOOLTIPS,title='APPL Stock Price',)
p.line('date','close',source=source)
show(p)
```
