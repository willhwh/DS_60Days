```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
```

```python
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
data
```

```python
#boxplot
sns.boxplot(data=data)
```

```python
#violinplot
sns.violinplot(data=data)
sns.despine(offset=10, trim=True);
#offset: away from the axes 
#trim: if True:  limit spines to the smallest and largest major tick on each non-despined axis.
```

```python
#function of sns.despine
sns.set_style("whitegrid")
sns.boxplot(data=data, palette="deep")
sns.despine(left=True)# if True: 删除左边边框
```

```python
# 散點圖關聯變量
tips = sns.load_dataset("tips")
sns.set(style="darkgrid")
sns.relplot(x="total_bill", y="tip", data=tips);
```

```python
#使用「色調語義」,因為點的顏色獲得意義:
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips);
```

```python
# 通過獨立地改變每個點的色調和样式，也可以表示四個變量(x,y,smoker,style)
sns.relplot(x="total_bill", y="tip", hue="smoker", style="time", data=tips);
```

```python
fmri = sns.load_dataset("fmri")
#透過replot 畫lineplot
#預設行為是透過繪製平均值和 95% 信心區間圍繞平均值來聚合每個值的多個測量值:
#自信區間是通過自助取樣法（bootstrapping）計算的
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri);
```

```python
#信心區間是使用引導計算的,對於較大的數據集,這需要時間。因此,可以禁用它們:

#ci: [int,'sd',None],置信区间的大小，当为‘sd’时绘制数据的标准差
sns.relplot(x="timepoint", y="signal", ci=None, kind="line", data=fmri);#ci = None
```

```python
#另一個不錯的選擇(尤其是對於較大的數據)是通過繪製標準差而不是信心區間來表示每個時間點的分佈分佈:
sns.relplot(x="timepoint", y="signal", kind="line", ci="sd", data=fmri);#ci = standard deviation 
```

```python
sns.relplot(x="timepoint", y="signal", hue="event", kind="line", data=fmri);
```

```python
sns.relplot(x="timepoint", y="signal", hue="region", style="event",
            kind="line", data=fmri);
```

```python
#單變數分佈圖
#默認情況下,這將繪製直方圖並適合內核密度估計值(KDE)。
x = np.random.normal(size=100)
sns.distplot(x);
```

```python
# bin: 指的是特徵值(直方條 總個數）, 先取消KDE 避免混淆
sns.distplot(x, bins=20, kde=False, rug=True);#rug:是否顯示資料分佈情況＃
```

```python
rs = np.random.RandomState(8)
x = rs.normal(2, 1, 75)
y = 2 + 1.5 * x + rs.normal(0, 2, 75)
```

```python
#模型殘差 圖型(residplot()): 该方法将在x上回归y，然后绘制残差的散点图。
sns.set_style('whitegrid')#有五種預設seaborn主題：darkgrid，whitegrid，dark，white
sns.residplot(x=x, y=y,color='red');
```

```python
#繪製單變數分佈:
sns.set_style('dark')
sns.distplot(x,kde=False,bins=10,rug=True)
```

```python
sns.set_style('dark')
sns.distplot(x,kde=False,bins=30,rug=False)
```

```python
sns.set_style('dark')
sns.distplot(y,bins=10)
```
