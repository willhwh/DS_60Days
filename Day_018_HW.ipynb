```python
#畫出 cos 圖檔，並儲存
x = np.arange(0,180)
y = np.cos(x)


plt.plot(y)
plt.show()

# 如果要存成圖形檔:
# 把 pyplot.show() 換成下面這行:plt.savefig("filename.png",dpi=300,format="png")
```
